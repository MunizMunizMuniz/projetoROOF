```python
import numpy as np
import pandas as pd 
from pyzipcode import ZipCodeDatabase
zcdb = ZipCodeDatabase()
```

```python
df = pd.read_csv('/content/kc_house_data.csv')
pd.set_option('display.max_columns', 50)
display(df)
df.describe()
```

Limpando valores nulos e duplicados

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

Transformando o dados da coluna date para tipo 'datetime'

```python
df['date'] = df.date.str[:8]
df['date'] = pd.to_datetime(df['date'])
```

Eliminando linha com 0 na coluna 'bedrooms'

```python
bed0 = df[df['bedrooms'] == 0].index
df = df.drop(bed0, axis=0)
```

Eliminando linha com 0 na coluna 'bathrooms'

```python
bath0 = df[df['bathrooms'] == 0].index
df = df.drop(bath0, axis=0)
```

Criando a coluna cidade a partir do zipcode

```python
cidade = []

for zipcode in df['zipcode']:
  try:
    zc = zcdb[zipcode]
    cidade.append(zc.city)
  except:
    cidade.append(0)
df.insert(loc= 3, column='city', value=cidade)
```

Limpando linhas da coluna 'city' com valores inválidos

```python
cidade0 = df[df['city'] == 0].index
df = df.drop(cidade0, axis=0)
```

Tirando colunas desnecessárias para análise

```python
df = df.drop('zipcode', axis=1)
df = df.drop('lat', axis=1)
df = df.drop('long', axis=1)
```

Eliminando linhas onde área habitável é maior que o tamanho do terreno


```python
areaCorrompida = df[df['sqft_living'] > df['sqft_lot']].index 
df = df.drop(areaCorrompida, axis=0)
```

Eliminando linha muito fora dos padrões, provavelmente corrompida

```python
bed33 = df[df['bedrooms'] == 33].index 
df = df.drop(bed33, axis=0)
```

Criando coluna de preço por área

```python
df['sqft_price'] = df['price'] / df['sqft_lot']
df['sqft_price_living'] = df['price'] / df['sqft_living']
```
